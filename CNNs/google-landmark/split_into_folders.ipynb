```python
file_list = os.listdir('data/val')
```

```python
labels = np.unique(data.values[:, 2])
```

```python
for label in tqdm_notebook(labels):
    os.mkdir('data/val/' + str(label))
```

```python
for cur_file in tqdm_notebook(file_list):
    file_class = data.values[:, 2][data.values[:, 0] == cur_file[:-4]][0]
    os.rename('/home/belozerova/google-landmark/data/val/' + cur_file, 
              '/home/belozerova/google-landmark/data/val/' + str(file_class) + '/' + cur_file)
```
